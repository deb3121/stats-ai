```python
import pandas as pd 
from sklearn import linear_model
```

```python
df=pd.read_csv('homeprices_2.csv') #dataframe for calling the function
```

```python
df
```

```python
df.bedrooms.median() #for calling median of the column 
```

```python
df.bedrooms=df.bedrooms.fillna(df.bedrooms.median())   #fill the median 
```

```python
df  #preprocessing the data 
```

```python
x=df.drop('price',axis=1)  #cutting dataframe , by default it will cut on rows and we have to mention the axis 
```

```python
x
```

```python
y=df.price
```

```python
y
```

```python
reg=linear_model.LinearRegression()  #building the ML model and using the package
```

```python
reg.fit(x,y)  #fit is a method and we are passing independent and dependent variable 
```

```python
reg.coef_  #predefined variable name 
```

```python
reg.intercept_   #for displaying b
```

```python
reg.predict([[3000,3,40]]) #solution for problem 1 
```
