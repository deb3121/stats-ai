```python
import pandas as pd   #dataframe
from sklearn import linear_model  #for linear regression
import matplotlib.pyplot as plt #for drawing graph
df=pd.read_csv('homeprices.csv')
df
#model will be made through data - ml model
```

```python
df.area #for showing area
df.price #for showing price
```

```python
plt.xlabel('AREA')  #for labelling x axis
plt.ylabel('PRICE')  #for labelling y axis
plt.plot(df.area,df.price,marker='*',color='red') #pass 2 values 1 is area and 1 is price 
plt.show() #plot will be connecting the dots 
```

```python
# x is area and it is independent variable 
x=df.drop('price' , axis='columns') #drop the price model then remaining part will be area now we will split the dataframe and by default it will be rows. 
x
```

```python
# y is price
y=df.price
y
```

```python
#last step is to develop the ml model 
reg=linear_model.LinearRegression() #package name - package means it can have n number of programs and each program has a specific purpose - linear_model = building linear regression ml model
reg.fit(x,y)  #passing x and y
```

```python
#finding m and b 
reg.coef_   #value of m 
```

```python
reg.intercept_  #value of b 
```

# y=mx+b

# price=coeff*area+intercept

```python
price=reg.coef_*3300+reg.intercept_
price
```

```python
area=int(input("Enter the area: "))  #dynamic model for finding price of area 
price=reg.coef_*area+reg.intercept_   
price
```
