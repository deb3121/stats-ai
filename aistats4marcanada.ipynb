```python
import pandas as pd 
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
plt.xlabel('YEAR')
plt.ylabel('PER CAPITA INCOME')
plt.scatter(df.year,df.per_capita_income,marker='*',color ="red")
plt.show()
```

```python
plt.plot(df.year,df.per_capita_income,color ="red")
plt.show()
```

```python
x = df.drop('per_capita_income', axis='columns')
x
```

```python
y= df.per_capita_income
y
```

```python
reg = linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
income = reg.coef_ * 2020 + reg.intercept_
income
```

```python
reg.predict([[2020]])
```
